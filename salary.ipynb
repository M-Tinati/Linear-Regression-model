```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("Salary_dataset.csv")
df = df.drop(columns=[df.columns[0]])
df.head()
```

```python
years = df["YearsExperience"].to_numpy().reshape(-1,1)
salary = df["Salary"].to_numpy()
```

```python
plt.scatter(years, salary, label='Actual Data', color='blue')
plt.grid(True , color='grey')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

```

```python
model = LinearRegression()
model.fit(years , salary)
```

```python
newyear = np.array([[13]])
predicted_year = model.predict(newyear)
```

```python
def result():
    
    plt.scatter(years, salary, label='Actual Data', color='blue')
    plt.scatter(newyear, predicted_year, color='red', label='New Years')
    plt.text(newyear, predicted_year, f'  ${predicted_year[0]:,.2f}', color='black')
    plt.xlabel('Years of Experience')
    plt.ylabel('Salary')
    plt.grid(True, color="grey")
    plt.legend()
    plt.show()
result()
```
